```python
import torch
from torch.utils.data import Dataset
import os
import cv2
import glob
import pandas as pd
import xml.etree.ElementTree as ET
import albumentations as A
from albumentations.pytorch import ToTensorV2
#from bounding_box import bounding_box as bb
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
root_path="D:/Dataset/Pet_Data/"
img_path=os.path.join(root_path,"images")
annotation_path=os.path.join(root_path,"annotations/xmls")           
annots = glob.glob(annotation_path+"/*.xml")
```

```python
#Dataset Class
class PetData(Dataset):

    
    def __init__(self, dataframe,train=False,transform=None):
        self.df=dataframe
        if transform is not None:
            if train:
                self.transform=transform['train']
            else:
                self.transform=transform['valid']
    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        fn,target,xmin,ymin,xmax,ymax=df.iloc[idx]
        im=cv2.cvtColor(cv2.imread(fn),cv2.COLOR_BGR2RGB)

        class_label=list([target])
        boxes   = [[xmin,ymin,xmax,ymax]]
        transformed = self.transform(image=im, bboxes=boxes, class_labels=class_label)
        image=transformed["image"]
        changed_boxes = torch.tensor(transformed['bboxes'], dtype=torch.float32)
        class_labels = torch.tensor(transformed["class_labels"])
        
        return image,changed_boxes,class_label
    
#Change xml to dataframe
def xml_to_csv(pths):
    CLASS_NAME=['cat','dog']
    '''pths: list of xml_files'''
    xml_list = []
    for xml_file in pths:
        # Read in the xml file
        tree = ET.parse(xml_file)
        path = os.path.join(img_path, tree.findtext("./filename"))
        height = int(tree.findtext("./size/height"))
        width = int(tree.findtext("./size/width"))
        xmin = int(tree.findtext("./object/bndbox/xmin"))
        ymin = int(tree.findtext("./object/bndbox/ymin"))
        xmax = int(tree.findtext("./object/bndbox/xmax"))
        ymax = int(tree.findtext("./object/bndbox/ymax"))
        name = CLASS_NAME.index(tree.findtext("./object/name"))
        xml_list.append([path,name,xmin,ymin,xmax,ymax])
    col_n = ["filename", "target","xmin", "ymin", "xmax", "ymax"]
    df = pd.DataFrame(xml_list, columns=col_n)
    return df
def disp_batch(batch,batch_img=False) :
    " funtion to display batch"
    if batch_img:
        image, target, _ = batch
        n = np.random.randint(0, len(image))
        im = image[n].permute(1, 2, 0).numpy()
        bbox=target[n][0]
    else:
        image, target, _ = batch
        im=image.permute(1,2,0).numpy()
        bbox=target[0]
    

    xy=(bbox[0],bbox[1])
    width=bbox[2]-bbox[0]
    height=bbox[2]-bbox[0]
    fig, ax = plt.subplots()
    ax.imshow(np.array(im))
    ax.add_patch(
     patches.Rectangle(
        xy,
        width,
        height,
        edgecolor = 'blue',
        fill=False ) )
    
    plt.show()
```

```python
transformations = [

    A.Resize(256,256),
    A.RandomCrop(224,224),
    A.HorizontalFlip(p=0.5),
]

# Train Transformations
train_transformations = transformations + [ A.ToFloat(max_value=255), 
                                            ToTensorV2() ]

# Valid Transformations
valid_transformations = [ A.ToFloat(max_value=255., always_apply=True), 
                          ToTensorV2(always_apply=True) ]

# Transformations:
transforms = {
    "train":A.Compose(train_transformations, p=1.0, bbox_params=A.BboxParams(format="pascal_voc", 
                                                                             label_fields=["class_labels"])),
              
    "valid": A.Compose(valid_transformations, p=1.0, bbox_params=A.BboxParams(format="pascal_voc", 
                                                                              label_fields=["class_labels"]))
    }
```

```python
df=xml_to_csv(annots)
df.head()
```

```python
def collate_fn(batch):
    return tuple(zip(*batch))
```

```python
BATCH_SIZE = 20

train_ds = PetData(df, train=True,transform=transforms)
train_dl = torch.utils.data.DataLoader(train_ds, batch_size=BATCH_SIZE
                                       , collate_fn=collate_fn)
```
